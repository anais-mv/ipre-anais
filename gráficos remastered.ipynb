```python
import matplotlib.pyplot as plt
import numpy as np

mse = [0, 5, 10, 20, 100, 200]
```

```python
def obtener_array_string(lista):
    lista = lista.replace("[","").replace("]","").split(",")
    lista = [float(num) for num in lista]
    return np.array(lista)

def obtener_lista_archivo(nombre):
    archivo = open(nombre, "r")
    lista = archivo.readlines()
    archivo.close()
    nueva_lista = []
    for elemento in lista:
        nueva_lista.append(elemento.strip("\n"))
    return nueva_lista

def obtener_lista_grafico(lista_listas, indice):
    final = []
    for elemento in lista_listas:
        final.append(elemento[indice])
    return final

def obtener_datos(lista_archivos):
    mse = [0, 5, 10, 20, 100, 200]
    
    expansiones_a_0 = []
    expansiones_a_5 = []
    expansiones_a_10 = []
    expansiones_a_20 = []
    expansiones_a_100 = []
    expansiones_a_200 = [] 
    
    tiempo_a_0 = []
    tiempo_a_5 = []
    tiempo_a_10 = []
    tiempo_a_20 = []
    tiempo_a_100 = []
    tiempo_a_200 = []

    expansiones_fs_0 = []
    expansiones_fs_5 = []
    expansiones_fs_10 = []
    expansiones_fs_20 = []
    expansiones_fs_100 = []
    expansiones_fs_200 = []

    tiempo_fs_0 = []
    tiempo_fs_5 = []
    tiempo_fs_10 = []
    tiempo_fs_20 = []
    tiempo_fs_100 = []
    tiempo_fs_200 = []

    expansiones_fsd_0 = []
    expansiones_fsd_5 = []
    expansiones_fsd_10 = []
    expansiones_fsd_20 = []
    expansiones_fsd_100 = []
    expansiones_fsd_200 = []

    tiempo_fsd_0 = []
    tiempo_fsd_5 = []
    tiempo_fsd_10 = []
    tiempo_fsd_20 = []
    tiempo_fsd_100 = []
    tiempo_fsd_200 = []
    
    for nombre in lista_archivos:
        archivo = obtener_lista_archivo(nombre)
        for linea in archivo:
            if "Nodos Focal:" in linea:
                lista = linea.replace("Nodos Focal: ", "")
                lista = obtener_array_string(lista)
                expansiones_fs_0.append(lista[0])
                expansiones_fs_5.append(lista[1])
                expansiones_fs_10.append(lista[2])
                expansiones_fs_20.append(lista[3])
                expansiones_fs_100.append(lista[4])
                expansiones_fs_200.append(lista[5])
            elif "Nodos Focal Discrepancy:" in linea:
                lista = linea.replace("Nodos Focal Discrepancy: ", "")
                lista = obtener_array_string(lista)
                expansiones_fsd_0.append(lista[0])
                expansiones_fsd_5.append(lista[1])
                expansiones_fsd_10.append(lista[2])
                expansiones_fsd_20.append(lista[3])
                expansiones_fsd_100.append(lista[4])
                expansiones_fsd_200.append(lista[5])
            elif "Nodos Astar:" in linea:
                lista = linea.replace("Nodos Astar: ", "")
                lista = obtener_array_string(lista)
                expansiones_a_0.append(lista[0])
                expansiones_a_5.append(lista[1])
                expansiones_a_10.append(lista[2])
                expansiones_a_20.append(lista[3])
                expansiones_a_100.append(lista[4])
                expansiones_a_200.append(lista[5])
            elif "Tiempo Focal:" in linea:
                lista = linea.replace("Tiempo Focal: ", "")
                lista = obtener_array_string(lista)
                tiempo_fs_0.append(lista[0])
                tiempo_fs_5.append(lista[1])
                tiempo_fs_10.append(lista[2])
                tiempo_fs_20.append(lista[3])
                tiempo_fs_100.append(lista[4])
                tiempo_fs_200.append(lista[5])
            elif "Tiempo Focal Discrepancy:" in linea:
                lista = linea.replace("Tiempo Focal Discrepancy: ", "")
                lista = obtener_array_string(lista)
                tiempo_fsd_0.append(lista[0])
                tiempo_fsd_5.append(lista[1])
                tiempo_fsd_10.append(lista[2])
                tiempo_fsd_20.append(lista[3])
                tiempo_fsd_100.append(lista[4])
                tiempo_fsd_200.append(lista[5])
            elif "Tiempo Astar:" in linea:
                lista = linea.replace("Tiempo Astar: ", "")
                lista = obtener_array_string(lista)
                tiempo_a_0.append(lista[0])
                tiempo_a_5.append(lista[1])
                tiempo_a_10.append(lista[2])
                tiempo_a_20.append(lista[3])
                tiempo_a_100.append(lista[4])
                tiempo_a_200.append(lista[5])
            elif "Valores g:" in linea:
                lista = linea.replace("Valores g: ", "")
                lista = obtener_array_string(lista)
                valor_g = lista[0]
    tiempos_fs = [tiempo_fs_0, tiempo_fs_5, tiempo_fs_10, tiempo_fs_20, tiempo_fs_100, tiempo_fs_200]
    tiempos_fsd = [tiempo_fsd_0, tiempo_fsd_5, tiempo_fsd_10, tiempo_fsd_20, tiempo_fsd_100, tiempo_fsd_200]
    tiempos_a = [tiempo_a_0, tiempo_a_5, tiempo_a_10, tiempo_a_20, tiempo_a_100, tiempo_a_200]
    expansiones_fs = [expansiones_fs_0, expansiones_fs_5, expansiones_fs_10, expansiones_fs_20, expansiones_fs_100, expansiones_fs_200]
    expansiones_fsd = [expansiones_fsd_0, expansiones_fsd_5, expansiones_fsd_10, expansiones_fsd_20, expansiones_fsd_100, expansiones_fsd_200]
    expansiones_a = [expansiones_a_0, expansiones_a_5, expansiones_a_10, expansiones_a_20, expansiones_a_100, expansiones_a_200]
    return (tiempos_a, tiempos_fs, tiempos_fsd, expansiones_a, expansiones_fs, expansiones_fsd, valor_g)
```

```python
def graficar(archivo, exp):
    archivo = "datos//" + archivo
    lista_archivos = [archivo + str(i) + ".txt" for i in range(0, 30)]

    tiempos_a, tiempos_fs, tiempos_fsd, expansiones_a, expansiones_fs, expansiones_fsd, valor_g = obtener_datos(lista_archivos)
    promedio_tiempo_a = [np.mean(tiempo) for tiempo in tiempos_a]
    promedio_tiempo_fs = [np.mean(tiempo) for tiempo in tiempos_fs]
    promedio_tiempo_fds = [np.mean(tiempo) for tiempo in tiempos_fsd]
    promedio_exp_a = [np.mean(exp) for exp in expansiones_a]
    promedio_exp_fs = [np.mean(exp) for exp in expansiones_fs]
    promedio_exp_fds = [np.mean(exp) for exp in expansiones_fsd]
    valores_g = [valor_g, valor_g, valor_g, valor_g, valor_g, valor_g]
    
    # gr치fico tiempo vs mse
    fig,ax = plt.subplots(dpi=144, figsize = [6,4])
    ax.plot([1,2,3,4,5,6], promedio_tiempo_fs, label="FS", c="red")
    ax.plot([1,2,3,4,5,6], promedio_tiempo_fds, label="FDS", c="blue")
    # ax.plot([1,2,3,4,5,6], promedio_tiempo_a, label="A*", c="purple")
    ax.set_xticks([1,2,3,4,5,6], mse)
    ax.legend(fontsize=14)
    ax.set_xlabel("Heuristic MSE", fontsize=14)
    ax.set_ylabel("Runtime (s)", fontsize=14)
    ax.set_title(f"Mean Time vs MSE - exp: {exp} - W: {w}", fontsize=14)
    # ax.set_ylim(1,4)
    plt.yscale("log")
    
    # gr치fico expansiones vs mse
    fig,ax = plt.subplots(dpi=144, figsize = [6,4])
    ax.plot([1,2,3,4,5,6], promedio_exp_fs, label="FS", c="red")
    ax.plot([1,2,3,4,5,6], promedio_exp_fds, label="FDS", c="blue")
    ax.plot([1,2,3,4,5,6], promedio_exp_a, label="A*", c="purple")
    # ax.plot([1,2,3,4,5,6], valores_g, label="g", c="orange", ls="--")
    ax.set_xticks([1,2,3,4,5,6], mse)
    ax.legend(fontsize=14)
    ax.set_xlabel("Heuristic MSE", fontsize=14)
    ax.set_ylabel("Expansions", fontsize=14)
    ax.set_title(f"Mean Expansions vs MSE - exp: {exp} - W: {w}", fontsize=14)
    # ax.set_ylim(25,35)
    plt.yscale("log")
```

```python
def graficar_mediana(archivo, exp):
    archivo = "datos//" + archivo
    lista_archivos = [archivo + str(i) + ".txt" for i in range(0, 30)]

    tiempos_a, tiempos_fs, tiempos_fsd, expansiones_a, expansiones_fs, expansiones_fsd, valor_g = obtener_datos(lista_archivos)
    promedio_tiempo_a = [np.median(tiempo) for tiempo in tiempos_a]
    promedio_tiempo_fs = [np.median(tiempo) for tiempo in tiempos_fs]
    promedio_tiempo_fds = [np.median(tiempo) for tiempo in tiempos_fsd]
    promedio_exp_a = [np.median(exp) for exp in expansiones_a]
    promedio_exp_fs = [np.median(exp) for exp in expansiones_fs]
    promedio_exp_fds = [np.median(exp) for exp in expansiones_fsd]
    valores_g = [valor_g, valor_g, valor_g, valor_g, valor_g, valor_g]
    
    # gr치fico tiempo vs mse
    fig,ax = plt.subplots(dpi=144, figsize = [6,4])
    ax.plot([1,2,3,4,5,6], promedio_tiempo_fs, label="FS", c="red")
    ax.plot([1,2,3,4,5,6], promedio_tiempo_fds, label="FDS", c="blue")
    # ax.plot([1,2,3,4,5,6], promedio_tiempo_a, label="A*", c="purple")
    ax.set_xticks([1,2,3,4,5,6], mse)
    ax.legend(fontsize=14)
    ax.set_xlabel("Heuristic MSE", fontsize=14)
    ax.set_ylabel("Runtime (s)", fontsize=14)
    ax.set_title(f"Median Time vs MSE - exp: {exp}", fontsize=14)
    # ax.set_ylim(1,4)
    plt.yscale("log")
    
    # gr치fico expansiones vs mse
    fig,ax = plt.subplots(dpi=144, figsize = [6,4])
    ax.plot([1,2,3,4,5,6], promedio_exp_fs, label="FS", c="red")
    ax.plot([1,2,3,4,5,6], promedio_exp_fds, label="FDS", c="blue")
    ax.plot([1,2,3,4,5,6], promedio_exp_a, label="A*", c="purple")
    # ax.plot([1,2,3,4,5,6], valores_g, label="g", c="orange", ls="--")
    ax.set_xticks([1,2,3,4,5,6], mse)
    ax.legend(fontsize=14)
    ax.set_xlabel("Heuristic MSE", fontsize=14)
    ax.set_ylabel("Expansions", fontsize=14)
    ax.set_title(f"Median Expansions vs MSE - exp: {exp}", fontsize=14)
    # ax.set_ylim(25,35)
    plt.yscale("log")
```
