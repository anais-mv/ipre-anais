```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
from funciones_graficos_mse import cargar, graficar_mse, lista_de_lista, eliminar_vacias

mse_values = [0, 5, 10, 20, 100, 200]
```

```python
fig,ax = plt.subplots(dpi=144, figsize = [10,4])
for i in range(1, 6):
    mse_k2 = lista_de_lista(100)
    for estado in grafo1.heuristics_k2[0]:
        h_perfect = grafo1.heuristics_k2[0][estado]
        indice = int(h_perfect)
        h_nn = grafo1.heuristics_k2[i][estado]
        # err_k2 = (h_nn - h_perfect)**2
        err_k2 = np.abs(h_nn - h_perfect)
        mse_k2[indice].append(err_k2)
    mse_k2 = eliminar_vacias(mse_k2)
    mean_mse_k2 = [np.mean(mse) for mse in mse_k2]
    h_perfect = [i for i in range(0, len(mean_mse_k2))]
    ax.plot(h_perfect, mean_mse_k2, label=f"MSE = {mse_values[i]}")
ax.set_title("k = 2", size=20)
ax.set_xlabel(r"$h*$", size=20)
ax.set_ylabel(r"$\overline{(h_{nn} - h*)^2}$", size=20)
# ax.plot([400 for i in range(0, 26)])
# ax.set_ylim(0, 2000)
ax.legend(fontsize=15)

len_mse = [len(mse) for mse in mse_k2]
```

```python
i = 1 # MSE = 5
k = 2
mse_k2 = lista_de_lista(100)
modelo = []
sum_h = 0
for estado in grafo3.heuristics_k2[0]:
    h_perfect = grafo3.heuristics_k2[0][estado]
    indice = int(h_perfect)
    h_nn = grafo3.heuristics_k2[i][estado]
    err_k2 = np.abs(h_nn - h_perfect)
    mse_k2[indice].append(err_k2)
    sum_h += h_perfect**(2*k)
mse_k2 = eliminar_vacias(mse_k2)
mean_mse_k2 = [np.mean(mse) for mse in mse_k2]
h_perfect = [i for i in range(0, len(mean_mse_k2))]

# MODELO
sum_h = sum_h/len(grafo3.heuristics_k2[0])
c = np.sqrt(mse_values[i]/sum_h)
for h in h_perfect:
    modelo.append(c * (h)**k)

# PLOT
fig, ax = plt.subplots(dpi = 144, figsize = [10, 4])
ax.plot(h_perfect, mean_mse_k2, label="exp")
ax.plot(h_perfect, modelo, label="model")
ax.set_title(f"MSE = {mse_values[i]} - k = {k}", size=20)
ax.set_xlabel(r"$h*$", size=20)
ax.set_ylabel(r"$\overline{(h_{nn} - h*)}$", size=20)
ax.legend(fontsize=15)
```
